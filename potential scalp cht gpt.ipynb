```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# --- Define your universe ---
tickers = ["NVDA","MSFT","AAPL","AMZN","META","AVGO","GOOGL","TSLA","GOOG","JPM","LLY","BRK-B","V","XOM","NFLX","MA","JNJ","ORCL","WMT","COST","ABBV","PLTR","HD","AMD","PG","BAC","UNH","GE","CVX","WFC","CSCO","IBM","KO","PM","CAT","RTX","CRM","GS","MU","MCD","MRK","ABT","TMO","LIN","PEP","DIS","UBER","NOW","MS","ISRG","LRCX","QCOM","INTU","AMAT","C","T","AXP","NEE","BKNG","BA","APH","SCHW","GEV","BLK","TJX","AMGN","TXN","KLAC","BSX","SPGI","VZ","GILD","ANET","INTC","ADBE","DHR","APP","PANW","HON","PFE","COF","LOW","SYK","UNP","PGR","CRWD","ADI","WELL","PLD","DE","LMT","ADP","COP","CEG","BX","VRTX","CMCSA","MO","SO","TMUS","CVS","HOOD","DUK","MCK","CME","NEM","PH","SBUX","MMC","MMM","CDNS","ICE","BMY","AMT","DASH","NOC","WM","ORLY","KKR","NKE","CI","EQIX","GD","MDLZ","HWM","ELV","RBLX","SHW","BK","RCL","HCA","MSI","TDG","EMR","SNOW","WMB","MSTR","USB","JCI","AJG","MCO","PNC","ITW","MRVL","ECL","FI","GLW","CTAS","CSX","ADSK","AZO","COIN","PYPL","GM","NSC","UPS","VST","CL","HLT","NET","PWR","AEP","URI","COR","ALNY","FCX","REGN","MPC","SRE","MAR","TRV","ZTS","EOG","DLR","AFL","CMI","TFC","SPG","CMG","APD","LHX","PSX","O","FTNT","VLO","BDX","SLB","AXON","D","WDAY","PCAR","KMI","FDX","ROP","ALL","IDXX","EA","ROST","APO","FAST","WBD","MPWR","ABNB","DDOG","PSA","F","MNST","EXC","CBRE","BKR","LNG","STX","FERG","DELL","CPNG","XEL","TTWO","CARR","AIG","AMP","RRG","FLTR","EW","GWW","MET","OKE","WDC","EBAY","VEEV","AME","CTVA","NDAQ","CVNA","TGT","XYZ","CCI","DHI","KR","A","PEG","ETR","YUM","PAYX","MSCI","ROK","CPRT","KMB","VMC","RMD","FICO","CAH","SYY","MLM","WEC","ED","OTIS","PCG","PRU","IQV","XYL","KDP","FIS","HIG","HUM","MCHP","GEHC","CTSH","TRGP","WAB","VICI","VRSK","EQT","CSGP","IR","EXR","VTR","NUE","SOFI","NRG","ZS","OXY","CCL","TSCO","STT","HPE","IRM","EME","DD","ARES","INSM","ADM","RJF","KVUE","MTB","AEE","EFX","MTD","KEYS","FIX","PSTG","LEN","RKLB","FANG","AWK","ATO","FITB","IBKR","DXCM","BR","SYF","ES","LPLA","HSY","PPL","AVB","EXPE","GIS","ODFL","HPQ","MDB","DTE","CNP","FE","BRO","IP","TDY","SMCI","VLTO","CBOE","NTRA","LDOS","HBAN","LYV","PTC","DOV","HUBS","EXE","CIEN","CINF","TTD","TPR","ULTA","STE","LH","PHM","NTAP","FLEX","EQR","K","PPG","MKL","RDDT","NTRS","FSLR","TER","PODD","HUBB","DG","WSM","TYL","WST","CRDO","EIX","CMS","TROW","BIIB","DRI","WRB","RF","CHTR","ON","CHD","DVN","ALAB","KHC","NVR","JBL","GPN","STLD","GWRE","BE","WAT","SBAC","CW","EL","HAL","CDW","CFG","NI","FWONK","DGX","ZBH","CASY","VRSN","DLTR","ZM","STZ","CPAY","KEY","J","TRMB","UTHR","GDDY","PKG","AFRM","PINS","IT","GPC","TOST","COHR","BWXT","NTNX","TPL","THC","PFG","SNA","TLN","SSNC","LNT","FTAI","CTRA","EVRG","ACM","USFD","FFIV","LVS","L","MKC","WY","ESS","SGI","ROL","DOW","BURL","FTV","IFF","INVH","CNC","RKT","DKNG","HOLX","EXPD","TRU","FCNCA","BBY","LII","BG","OMC","MAA","PFGC","ZBRA","SUI","IONQ","TWLO","ILMN","WWD","DECK","DT","DKS","FNF","CG","KIM","CHRW","JLL","COO","KTOS","GEN","TSN","WPC","EQH","MAS","OKTA","CSL","CLX","INCY","AVY","ASTS","OKLO","XPO","CRS","TXT","DOCU","RPM","NBIX","RS","EWBC","CF","BALL","AVAV","ENTG","RPRX","ITT","NLY","HEI","BLDR","ARE","LECO","TOL","GGG","DPZ","MEDP","REG","DOC","CACI","NXT","BLD","TW","RL","RGTI","APG","IEX","Z","EHC","UNM","JBHT","ROKU","RGA","RBC","MTZ","WSO","BJ","ATI","GLPI","TKO","BAH","VTRS","UDR","RGLD","CELH","NDSN","DUOL","LYB","OHI","EXAS","SCCO","BAX","TXRH","WTRG","MANH","CLH","BXP","HST","EVR","SF","CCK","CDE","FOXA","JKHY","SNX","MLI","UHS","RVTY","SCI","ALB","ALLY","HII","SWKS","LITE","IONS","U","AYI","PNW","CPT","FDS","DTM","GNRC","OC","SWK","GL","ELS","WMS","DAY","AKAM","HLI","SFM","RIVN","AIZ","SJM","AMH","LAMR","ELAN","AA","ENSG","ARMK","FHN","PAYC","AGNC","NCLH"]

def calculate_atr(df, period=14):
    """Calculate Average True Range for volatility measurement"""
    high_low = df['High'] - df['Low']
    high_close = np.abs(df['High'] - df['Close'].shift())
    low_close = np.abs(df['Low'] - df['Close'].shift())
    
    ranges = pd.concat([high_low, high_close, low_close], axis=1)
    true_range = np.max(ranges, axis=1)
    return true_range.rolling(period).mean()

def calculate_momentum_score(df, window=4):
    """Calculate recent momentum strength (last 20 mins = 4 bars of 5min)"""
    if len(df) < window:
        return 0
    recent_bars = df.tail(window)
    price_change = (recent_bars['Close'].iloc[-1] - recent_bars['Close'].iloc[0]) / recent_bars['Close'].iloc[0] * 100
    return price_change

def calculate_proper_rsi(series, period=14):
    """Fixed RSI calculation"""
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# --- Download Data (2 days to ensure enough historical data) ---
print("Downloading data...")
data = yf.download(tickers, period='2d', interval='5m', group_by='ticker', progress=False)

results = []

for t in tickers:
    try:
        df = data[t].dropna()
        if df.empty or len(df) < 20:
            continue
        
        # Calculate indicators
        df['EMA9'] = df['Close'].ewm(span=9, adjust=False).mean()
        df['EMA20'] = df['Close'].ewm(span=20, adjust=False).mean()
        df['VWAP'] = (df['Close'] * df['Volume']).cumsum() / df['Volume'].cumsum()
        df['RSI'] = calculate_proper_rsi(df['Close'], 14)
        df['ATR'] = calculate_atr(df)
        
        # Volume profile
        df['VolMA'] = df['Volume'].rolling(window=20).mean()
        
        latest = df.iloc[-1]
        prev = df.iloc[-2]
        
        # Skip if missing critical data
        if pd.isna(latest['RSI']) or pd.isna(latest['ATR']):
            continue
        
        # --- Enhanced Scalp Filters ---
        
        # 1. Trend confirmation: EMA alignment
        ema_bullish = latest['EMA9'] > latest['EMA20']
        
        # 2. Price above VWAP (institutional support)
        above_vwap = latest['Close'] > latest['VWAP']
        
        # 3. RSI in sweet spot (momentum but not overbought)
        rsi_ok = 45 < latest['RSI'] < 70
        
        # 4. Recent momentum (last 20 minutes positive)
        momentum_20min = calculate_momentum_score(df, window=4)
        positive_momentum = momentum_20min > 0
        
        # 5. Volume confirmation
        avg_vol = df['Volume'].tail(20).mean()
        rel_vol = latest['Volume'] / avg_vol if avg_vol > 0 else 0
        volume_surge = rel_vol >= 1.5  # Lowered from 2 for more opportunities
        
        # 6. Volatility check (ATR relative to price for scalping potential)
        atr_pct = (latest['ATR'] / latest['Close']) * 100
        sufficient_volatility = atr_pct > 0.3  # At least 0.3% ATR for scalping
        
        # 7. Price action: current bar showing strength
        bar_strength = (latest['Close'] - latest['Open']) / latest['Open'] * 100
        
        # 8. Calculate intraday metrics
        today_open = df[df.index.date == df.index[-1].date()].iloc[0]['Open']
        change_pct = (latest['Close'] - today_open) / today_open * 100
        
        # 9. Spread check (High-Low range for entry precision)
        bar_range_pct = ((latest['High'] - latest['Low']) / latest['Low']) * 100
        
        # Apply all filters
        if (ema_bullish and above_vwap and rsi_ok and positive_momentum 
            and volume_surge and sufficient_volatility):
            
            # Additional quality filters
            if 0.5 <= change_pct <= 15:  # Reasonable intraday move
                
                # Calculate risk-reward metrics
                entry_price = latest['Close']
                stop_loss = entry_price - (1.5 * latest['ATR'])  # 1.5 ATR stop
                target_price = entry_price + (2 * latest['ATR'])  # 2:1 R/R minimum
                
                risk_pct = ((entry_price - stop_loss) / entry_price) * 100
                reward_pct = ((target_price - entry_price) / entry_price) * 100
                risk_reward = reward_pct / risk_pct if risk_pct > 0 else 0
                
                results.append({
                    'Ticker': t,
                    'Price': round(latest['Close'], 2),
                    'Change %': round(change_pct, 2),
                    '20min Mom': round(momentum_20min, 2),
                    'RSI': round(latest['RSI'], 1),
                    'Rel Vol': round(rel_vol, 2),
                    'ATR %': round(atr_pct, 2),
                    'Entry': round(entry_price, 2),
                    'Stop': round(stop_loss, 2),
                    'Target': round(target_price, 2),
                    'R:R': round(risk_reward, 2),
                    'Risk %': round(risk_pct, 2)
                })
    
    except Exception as e:
        # Skip tickers with errors (bad symbols, no data, etc.)
        continue

# --- Show results with beautiful formatting ---
if results:
    scalp_df = pd.DataFrame(results)
    # Sort by combination of momentum and risk-reward
    scalp_df['Score'] = scalp_df['20min Mom'] * scalp_df['R:R']
    scalp_df = scalp_df.sort_values(by='Score', ascending=False)
    scalp_df = scalp_df.drop('Score', axis=1)
    
    # Format output
    print("\n" + "="*100)
    print("üéØ SCALPING OPPORTUNITIES - WLS INDEX".center(100))
    print("="*100)
    print(f"Scan Time: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')} | Market: {'OPEN' if 9.5 <= datetime.now().hour + datetime.now().minute/60 <= 16 else 'CLOSED'}")
    print("-"*100)
    
    # Set pandas display options for better formatting
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_colwidth', None)
    
    # Color code based on momentum strength
    print("\nüìä TOP CANDIDATES (Sorted by Quality Score: Momentum √ó Risk:Reward)\n")
    
    for idx, row in scalp_df.head(10).iterrows():
        momentum_emoji = "üöÄ" if row['20min Mom'] > 1.5 else "üìà" if row['20min Mom'] > 0.5 else "‚û°Ô∏è"
        rr_emoji = "‚≠ê" if row['R:R'] > 2.5 else "‚úì"
        
        print(f"{momentum_emoji} {row['Ticker']:6} | ${row['Price']:8.2f} | Day: {row['Change %']:+6.2f}% | 20min: {row['20min Mom']:+5.2f}% | RSI: {row['RSI']:5.1f}")
        print(f"   Vol: {row['Rel Vol']:.2f}x | ATR: {row['ATR %']:.2f}% | Entry: ${row['Entry']:.2f} ‚Üí Stop: ${row['Stop']:.2f} ‚Üí Target: ${row['Target']:.2f}")
        print(f"   {rr_emoji} Risk: {row['Risk %']:.2f}% | Reward: {row['Risk %'] * row['R:R']:.2f}% | R:R = {row['R:R']:.2f}:1")
        print("-"*100)
    
    if len(scalp_df) > 10:
        print(f"\n... and {len(scalp_df) - 10} more candidates\n")
    
    # Summary statistics
    print("\n" + "="*100)
    print("üìà SUMMARY STATISTICS")
    print("="*100)
    print(f"Total Candidates Found: {len(scalp_df)}")
    print(f"Average 20min Momentum: {scalp_df['20min Mom'].mean():.2f}%")
    print(f"Average Risk:Reward: {scalp_df['R:R'].mean():.2f}:1")
    print(f"Average Relative Volume: {scalp_df['Rel Vol'].mean():.2f}x")
    print(f"Best Momentum: {scalp_df['Ticker'].iloc[scalp_df['20min Mom'].idxmax()]} ({scalp_df['20min Mom'].max():.2f}%)")
    print(f"Best R:R: {scalp_df['Ticker'].iloc[scalp_df['R:R'].idxmax()]} ({scalp_df['R:R'].max():.2f}:1)")
    print("="*100)
    
    # Export option
    filename = f"scalp_candidates_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
    scalp_df.to_csv(filename, index=False)
    print(f"\nüíæ Results saved to: {filename}")
    
else:
    print("\n" + "="*100)
    print("‚ùå NO SCALP CANDIDATES FOUND")
    print("="*100)
    print("\nPossible reasons:")
    print("  ‚Ä¢ Market is closed or low volume period")
    print("  ‚Ä¢ No stocks meet all criteria (try relaxing filters)")
    print("  ‚Ä¢ Waiting for setup to develop")
    print("\nüí° Tip: Run this during market hours (9:30 AM - 4:00 PM ET) for best results")
    print("="*100)
```
